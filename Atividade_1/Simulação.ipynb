```python
import subprocess
import os
import pandas as pd

def convert_to_mili(df):


    # Selecionar apenas as linhas até 1300
    df_subset = df.iloc[:1300]

    # Converter a coluna "delta" para um formato de tempo
    df_subset['delta'] = pd.to_timedelta(df_subset['delta'])

    # Converter para milissegundos
    df_subset['delta'] = df_subset['delta'].dt.total_seconds() * 1000


    return df_subset


def executar():
    csvFiles = [csvFile for csvFile in os.listdir(
    ) if 'csv' in csvFile and 'collectSample' not in csvFile]
    print(csvFiles)

    data_estats = {"file_name":[],"mean":[],"std_deviation":[]}

    for csvFile in csvFiles:
        df = pd.read_csv(csvFile)

        df = convert_to_mili(df)

        data_estats["file_name"].append(csvFile)
        data_estats["mean"].append(df["delta"].mean())
        data_estats["std_deviation"].append(df["delta"].std())

    df = pd.DataFrame(data = data_estats)

    df.to_csv("dados_atividade.csv",index = False)


```
